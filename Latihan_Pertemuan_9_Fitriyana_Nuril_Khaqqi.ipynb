FUNGSI ALGORITMA PAGERANK

```python
#untuk menghitung page rank
import numpy as np
def PAGE_RANK(matriks):
    #define the adjaceny matrix
    adjacency_matrix = np.array(matriks)

    #Set the damping factor
    damping_factor = 0.85

    #initialize the page rank value

    num_pages = len(adjacency_matrix)
    page_rank = np.ones(num_pages) / num_pages

    #Number of iterations
    num_iterations = 100

    for i in range(num_iterations):
        new_page_rank = np.zeros(num_pages)
        for j in range(num_pages):
            #calculate the Pagerank contribution from each linking page
            linking_pages = [k for k in range(num_pages) if adjacency_matrix[k,j]==1]
            for linking_page in linking_pages:
                new_page_rank[j] += page_rank[linking_page]/sum(adjacency_matrix[linking_page, :])
            new_page_rank[j]=damping_factor * new_page_rank[j] + (1 - damping_factor)/ num_pages
        page_rank = new_page_rank

    #print the final Pagerank values
    for page, rank in enumerate(page_rank):
        print(f"PR(page {page+1})={rank:.3f}")
```

FUNGSI MEMBUAT GRAPHIC NODE

```python
#untuk membuat grafik
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def GAMBAR (matriks):
    adjacency_matrix = np.array(matriks)

    #create a directed graph
    G = nx.DiGraph()

    #add nodes
    num_pages = len(adjacency_matrix)
    G.add_nodes_from(range(num_pages))

    #add edges based on the adjaceny matrix
    for i in range(num_pages):
        for j in range(num_pages):
            if adjacency_matrix[i, j] == 1:
                G.add_edge(i, j)

    #Draw the graph
    pos = nx.spring_layout(G) #Layout the graph
    labels = {i: f"Page{i+1}"for i in range(num_pages)}
    nx.draw(G, pos, with_labels=True, labels=labels, node_size=1000, node_color='lightblue', font_size=10, font_color='black', font_weight='bold', arrowsize=20)
    plt.title("PageRank Example Graph")
    plt.show()
```

Jawaban : **Analisa penerapan algoritma PageRank dalam penyebaran informasi**

Penyebaran informasi adalah salah satu permasalahan sosial yang penting untuk dijawab. Informasi yang relevan dan penting harus dapat disebarkan secara luas agar dapat diketahui oleh masyarakat. Algoritma PageRank dapat digunakan untuk menjawab permasalahan ini dengan cara menentukan peringkat halaman web yang relevan dan penting.

Algoritma PageRank bekerja dengan menghitung jumlah dan kualitas tautan yang mengarah ke halaman tersebut. Halaman web yang memiliki lebih banyak tautan masuk dari halaman web yang memiliki peringkat tinggi akan memiliki peringkat yang lebih tinggi juga. Hal ini berarti bahwa halaman web yang memiliki peringkat tinggi lebih mungkin ditemukan oleh pengguna mesin pencari.

Penyebaran informasi menggunakan algoritma PageRank memiliki beberapa kelebihan, yaitu:

* **Efektif**

Algoritma PageRank dapat digunakan untuk menyebarkan informasi secara luas dan efektif. Hal ini karena algoritma ini menggunakan tautan sebagai indikator relevansi dan pentingnya informasi. Tautan adalah cara yang paling alami untuk menyebarkan informasi di internet.

* **Objektif**

Algoritma PageRank adalah algoritma yang objektif. Hal ini berarti bahwa peringkat halaman web ditentukan oleh jumlah dan kualitas tautan yang mengarah ke halaman tersebut, tanpa dipengaruhi oleh faktor-faktor lain, seperti bias atau kepentingan pribadi.

* **Akurat**

Algoritma PageRank dapat digunakan untuk menentukan peringkat halaman web yang akurat. Hal ini karena algoritma ini menggunakan data yang nyata, yaitu jumlah dan kualitas tautan yang mengarah ke halaman tersebut.

**Contoh penerapan algoritma PageRank dalam penyebaran informasi**

Pemerintah dapat menggunakan algoritma PageRank untuk menyebarkan informasi tentang kebijakan atau program-programnya kepada masyarakat. Informasi tersebut dapat dipublikasikan di situs web pemerintah atau di media sosial. Algoritma PageRank dapat digunakan untuk menentukan halaman-halaman web pemerintah yang paling banyak dikunjungi oleh masyarakat. Dengan demikian, pemerintah dapat mengetahui informasi apa saja yang paling relevan bagi masyarakat.

Media massa dapat menggunakan algoritma PageRank untuk menentukan berita atau artikel yang paling penting untuk dipublikasikan. Berita atau artikel yang memiliki peringkat tinggi dalam mesin pencari akan lebih mudah ditemukan oleh pembaca. Dengan demikian, media massa dapat menyajikan berita atau artikel yang paling relevan bagi pembaca.

Organisasi non-pemerintah dapat menggunakan algoritma PageRank untuk memetakan hubungan antar pelanggan atau pemasok. Hubungan tersebut dapat digunakan untuk meningkatkan layanan pelanggan atau untuk menjalin kerja sama yang lebih baik dengan pemasok.

**Kesimpulan**

Penerapan algoritma PageRank dalam penyebaran informasi dapat memberikan manfaat yang besar bagi masyarakat. Algoritma ini dapat digunakan untuk menyebarkan informasi secara luas, efektif, objektif, dan akurat.
